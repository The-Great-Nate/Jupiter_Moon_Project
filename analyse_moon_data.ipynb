# Analysis of the Data from Jupiter's Moons.

This is a notebook regarding the analysis of a database regarding Jupiter's moons. A package was created for the purpose of analysing this database, however the package was designed with versatility in mind with regards to other databases.

The aim of using the database is to estimate the mass of Jupiter.

## Reading the database

Below is the database that is to be analysed. The package created for the analysis is first imported, then the function `display_df()` was ran. This function outputs the database as a pandas dataframe.

```python
from data_analysis_package.jupiter import Moons
```

```python
jupiter_moons = Moons("jupiter.db", "moons")
jupiter_moons.display_df()
```

## The Quantity of Fields and Records in the Dataset
`data_info()` outputs no. of fields (columns), records (rows) and a list of the column names in dataset.

```python
jupiter_moons.display_data_info()
```

## How complete is the dataset.
It is good to be able to see how many missing value in the database there is.

With the `check_missing_values()` method, the amount of missing values in each column is output as a pandas series.

```python
jupiter_moons.check_missing_values()
```

From the output dataframe of the database, it can be observed that the **mass_kg** column has a significant quantity of missing values. with **mag** also containing a few missing values.
From observing this, 50/56 of the values in the **mass_kg** column are declared as missing values (refered to as _NaN_). This is $\approx$ 90% of the records in the dataset.

below the `return_complete_records()` method returns all of the rows with complete data (no _NaN_ values)

```python
jupiter_moons.return_complete_records()
```

It can be observed from running the `return_complete_records()` method that the majority of the moons in the *Galilean* group have complete data.

Within the package, there is also the method `filter_for_characteristic(column:str, characteristic:str)` where the two paramters `column` and `characteristic` take string values for their respective names. From running this function with `"group"` and `"Galilean"` as arguements, it can be seen that all _Galilean_ type moons are complete records.

The `filter_for_characteristic(column:str, characteristic:str)` method returns a filtered pandas data frame and shares the same functionality as an unfiltered pandas dataframe as demonstrated from the `display_df()` function used earlier.

```python
display(jupiter_moons.filter_for_characteristic("group", "Galilean"))
no_galilean_moons = len(jupiter_moons.filter_for_characteristic("group", "Galilean"))
print(f"The quantity of moons in the Galilean group: {no_galilean_moons}")
```

To deduce why the majority of the other moons do not have data obtained for their mass. For a fast overview of each numeric variable's distribution against each other, the `pairwise_plots()` method outputs this high-level overview. The `hue` parameter of this method plots each datapoint as a colour based off the categoric variable column specified. Refer to the legend for which category is assigned to which colour.

```python
jupiter_moons.pairwise_plots(hue = "group")
```

Several deductions can be made from this pairplot to deduce the trend that variables that enable Galilean type moons have data for mass. Barring the mass_kg column, 

```python
jupiter_moons.plot_relationship("period_days", "period_days", hue = "group")
```
