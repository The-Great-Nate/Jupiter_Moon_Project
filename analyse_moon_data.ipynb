## Finding the Mass of Jupiter.

Kepler's Third Law of planetary motion can be used to estimate the mass of Jupiter. The full relationship between the orbital period ($T$) squared to the cube of the length of the semi-major axis ($\alpha$) is,

\begin{equation}
    T^2 = \frac{4\pi^2}{GM}\alpha^3 \quad,
\end{equation}

where $M$ is the mass of the central body and $G$ is the gravitational constant $6.67 \times 10^-11 m^3kg^{-1}s^{-2}$.

The central body in this scanario is Jupiter itself ; therefore, _period_days_ and _distance_km_ would represent the orbital period and distance away with respect to Jupiter and $M$ would represent the mass of Jupiter.

From observing the database, both _period_days_ and _distance_km_ columns need to be converted to standard units. As the database is a pandas dataframe which can be obtained through the `get_df()` method, new columns can be created by applying the numerical operations onto the entirety of the column and stating the name of the new column through indexing. To reduce the quantity of lines of code used, the power operations (squaring period and cubing distance) is also performed on the same lines. 

These new columns are now named:

 **period_seconds_p2**

 **distance_m_p3**
.

However to save these new columns in the dataframe, the `set_df()` method is required to run and have the new database be passed in as an arguement.

```python
database = jupiter_moons.get_df()
database["period_seconds_p2"] = (database["period_days"] * 24 * 60 * 60)**2
database["distance_m_p3"] = (database["distance_km"] * 1000)**3
jupiter_moons.set_df(database)
jupiter_moons.get_df()
```

### Making a Regression Model between $T^2$ and $\alpha^3$

Since $T^2$ and $\alpha^3$ are directly proportional to each other, in an ideal case, measurements of both variables should enable a straight line of best fit to be created. A linear regression machine learning model can be used to obtain the mass $M$ of Jupiter.

To create the instance of the model for $T^2$ and $\alpha^3$ the `get_model()` method can be used which takes in arguements for the x variable (`col_1`) and the y (`col_2`). As Kepler's Third Law does not assume an intercept. `fit_intercept = False`.

```python
model_tuple = jupiter_moons.get_model(col_1 = "distance_m_p3", col_2 = "period_seconds_p2", fit_intercept = False)
```

The `get_model()` method returns a tuple which stores
| Index | Variable  |
|-------|-----------|
| 0     | model     |
| 1     | x_test    |
| 2     | y_test    |
| 3     | col_1     |
| 4     | col_2     |.

In summary, the `get_model()` uses methods and classes from the sklearn library. `model` is an object of `LinearRegression`. `x_test` and `y_test` are arrays of data to test the model created by the `model_selection.train_test_split()` method. `col_1` and `col_2` are simply just the parameters input into the `get_model()` function which come in handy when using the other methods created that require the model.

To obtain the gradient from the linear regression model, the `get_model_parameters()` method can be used to do so as this method returns the gradient and intercept of the linear regression model generated from the training data (which is not returned from `get_model()`). The gradient is required to obtain $M$

\begin{equation}
    M = \frac{4 \pi^2}{G \times gradient} \;.
\end{equation}

```python
gradient, intercept = jupiter_moons.get_model_parameters(model_tuple)
```

From the linear regression parameters generated, predicted values can be generated based off the `x_test` array which is a part of the model tuple. The prediction can be returned through inputting this tuple as an arguement in the `predict()` method in the package.

```python
predicted_data = jupiter_moons.predict(model_tuple)
```

The result from the regression model's predictions can be plotted with the `plot_prediction()` method with passing the model tuple and prediction into the method as an arguement. The package has a method dedicated to returning the mass of the central body from Kepler's Third Law known as `return_central_mass()`. From storing the gradient and mass value, they can be output to the notebook with `print()`

```python
jupiter_moons.plot_prediction(model_tuple, predicted_data)
mass = jupiter_moons.return_central_mass(gradient)
print(f"The gradient the model states is: {gradient}")
print(f"The mass calculated from the model is therefore: {mass}")
```

From the output, the mass of Jupiter is estimated to be $1.90847 \times 10^{27} kg$ which is $\approx$ $0.5$% above the literature value from NASA of $1.89813 \times 10^{27} kg$ [[3]](https://nssdc.gsfc.nasa.gov/planetary/factsheet/jupiterfact.html)

It is worth checking the effectiveness/validity of the model even with the overfitting measures taken. A useful visualisation would be to observe the *residuals*, the difference between the observed value of the orbital period and the predicted value from the model.

```python
jupiter_moons.plot_residuals(model_tuple, predicted_data)
```

For a numerical metric to determine effectiveness of the model, a metric known as the _coefficient of determination_ (or $R^2$ value) is a measure that determines how closely the model agrees with the data given. A value close to `1.0` is considered to be the best possible score (model 100% agrees with the data).

The average distance between the predicted values and the actual values can be determined by the _root mean squared error_.

Both these values is returned from the `output_model_worth()` method by passing the model tuple and prediction as arguements.

```python
r2, rmse = jupiter_moons.output_model_worth(model_tuple, predicted_data)
```

The percentage error of the obtained mass could be obtained from the root mean squared error / obtained mass $\times$ 100.

```python
print((rmse/mass)*100)
```

Not only the $R^2$ score is close to 1, the obtained percentage error is very small which also shows how effective this regression model is.

To obtain a more accurate value for the mass of Jupiter, database could contain infomation regarding all 95 moons of Jupiter as with more sample data, leads to a more precise and potentially accurate value to the literature.

The database could be standardised through introducing _weights_ to each moon (not to be mixed up with the force), where moons that have been studied extensively for longer could perhaps hold a larger weight. However if one valued the measurements of each moon equally, using weights would not be neccessary. While Callisto, Europa, Ganymede, and Io are said to be the most well-studied of Jupiter's moons, giving these records a higher weighting to the other moons would make the dataset's output statistics more decisive on 4 moons out of 56, which is not a good sample size or representation of the whole dataset.
